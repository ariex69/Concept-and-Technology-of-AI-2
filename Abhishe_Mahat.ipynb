```python
import numpy as np

# Given Arrays A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A^-1 = I (Identity Matrix)
A_inv = np.linalg.inv(A)
identity_matrix = np.dot(A, A_inv)
print("A * A^-1 = \n", identity_matrix)

# 2. Prove AB != BA (Matrix multiplication is not commutative)
AB = np.dot(A, B)
BA = np.dot(B, A)
print("AB = \n", AB)
print("BA = \n", BA)

# 3. Prove (AB)^T = B^T * A^T (Transpose property)
AB_T = np.transpose(AB)
B_T = np.transpose(B)
A_T = np.transpose(A)
B_T_A_T = np.dot(B_T, A_T)
print("(AB)^T = \n", AB_T)
print("B^T * A^T = \n", B_T_A_T)

# 4. Solve the system of linear equations using Inverse Method
# The system is:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Define matrix A (coefficients) and matrix B (constants)
A_eq = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

# Solve using inverse method (A^-1 * B)
A_inv_eq = np.linalg.inv(A_eq)
X = np.dot(A_inv_eq, B_eq)
print("Solution using Inverse Method (x, y, z):")
print(X)  # This will give us [x, y, z]

# 5. Solve using np.linalg.inv (already solved in previous step)
# The np.linalg.inv function is already used to find the inverse and solve for X
```
